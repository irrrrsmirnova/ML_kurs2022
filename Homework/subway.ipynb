# Задача регрессии

```python
df_regr = df
df_regr.drop('ticket_type_nm', axis = 1, inplace = True)
df_regr.head(10)
```

```python
from sklearn.model_selection import train_test_split

x_regr_train, x_regr_test, y_regr_train, y_regr_test = train_test_split(df_regr.drop(columns=['time_to_under']), df_regr[['time_to_under']], test_size=0.3, random_state = 22)
```

```python
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingRegressor
```

```python
mdl = HistGradientBoostingRegressor()
```

```python
len(x_regr_train[:100000])
```

```python
mdl.fit(x_regr_train, y_regr_train)
```

```python
regres_pred = mdl.predict(x_regr_test)
```

```python
from sklearn.metrics import r2_score, recall_score
```

```python
r2 = r2_score(y_regr_test, regres_pred)
r2
```

# Задача классификации

```python
df_clf = df
df_clf.head(10)
```

```python
df_clf['pass_dttm'] = df_clf['pass_dttm'].astype(int)
type(df_clf['pass_dttm'][1])
```

```python
df_clf.head(10)
```

```python
x_clf_train, x_clf_test, y_clf_train, y_clf_test = train_test_split(df_clf.drop(columns=['label']), df_regr[['label']], test_size=0.3, random_state = 22)
```

```python
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
clf = HistGradientBoostingClassifier()
```

```python
clf.fit(x_clf_train, y_clf_train)
```

```python
clf_pred = clf.predict(x_clf_test)
```

```python
rec = recall_score(y_clf_test, clf_pred, average='micro')
rec
```
